```python
import numpy as np
import matplotlib.pyplot as plt
from generate_data import postprocess
```

```python
POINT_SIZE = 10
SELECTED_POINT_SIZE = 160
LABEL_SIZE_2D = 30
TICK_SIZE_2D = 20
LABEL_SIZE_3D = 30
TICK_SIZE_3D = 20
```

```python
data = np.load("results/data_pc.npy")
d_a, d_b, T_2a, T_2b = data.T

data2 = postprocess(data)
d_a2, d_b2, T_2a2, T_2b2 = data2.T
```

```python
selected_points = [12, 520, 3546, 5100, 5959] # these are the indices
colors = ['yellow', 'purple', 'orange', 'red', 'lime']    
```

```python
# Subplot a

plt.scatter(T_2a, T_2b, s=POINT_SIZE, alpha=0.5)
plt.xlabel(r'$T_{2a}$', fontsize=LABEL_SIZE_2D)
plt.ylabel(r'$T_{2b}$', fontsize=LABEL_SIZE_2D)
plt.xticks(fontsize=TICK_SIZE_2D)
plt.yticks(fontsize=TICK_SIZE_2D)
plt.xlim(-10, 250)
plt.ylim(-10, 250)

plt.gcf().set_size_inches(8, 8)

for i, color in zip(selected_points, colors):
    plt.scatter(T_2a[i], T_2b[i], c=color, s=SELECTED_POINT_SIZE, marker = "D", edgecolors = "black", linewidth = 0.5)

plt.tight_layout(pad = 1.5)
plt.gcf().set_size_inches(8, 8)

plt.savefig("results/plot1a.png")
plt.show()
    
plt.clf()
```

```python
# Subplot b

plt.scatter(T_2a2, T_2b2, s=POINT_SIZE, alpha=0.5)
plt.xlabel(r'$T_{21}$', fontsize=LABEL_SIZE_2D)
plt.ylabel(r'$T_{22}$', fontsize=LABEL_SIZE_2D)
plt.xticks(fontsize=TICK_SIZE_2D)
plt.yticks(fontsize=TICK_SIZE_2D)
plt.xlim(-10, 250)
plt.ylim(-10, 250)

plt.gcf().set_size_inches(8, 8)

for i, color in zip(selected_points, colors):
    plt.scatter(T_2a2[i], T_2b2[i], c=color, s=SELECTED_POINT_SIZE, marker = "D", edgecolors = "black", linewidth = 0.5)

plt.axline((0, 0), slope = 1, alpha = 0.5, c = "gray", linestyle = "--")

plt.tight_layout(pad = 1.5)
plt.gcf().set_size_inches(8, 8)
    
plt.savefig("results/plot1b.png")
plt.show()
    
plt.clf()
```

```python
# Subplot c

plt.scatter(d_a, d_b, s=POINT_SIZE, alpha=0.5)
plt.xlabel(r'$d_{a}$', fontsize=LABEL_SIZE_2D)
plt.ylabel(r'$d_{b}$', fontsize=LABEL_SIZE_2D)
plt.xticks(fontsize=TICK_SIZE_2D, rotation = 15)
plt.yticks(fontsize=TICK_SIZE_2D)

plt.gcf().set_size_inches(8, 8)

for i, color in zip(selected_points, colors):
    plt.scatter(d_a[i], d_b[i], c=color, s=SELECTED_POINT_SIZE, marker = "D", edgecolors = "black", linewidth = 0.5) 

plt.tight_layout(pad = 1.5)
plt.gcf().set_size_inches(9, 8)

plt.savefig("results/plot1c.png")
plt.show()
    
plt.clf()
```

```python
# Subplot d

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d', computed_zorder=False)
ax.scatter(d_a, T_2a, T_2b, s=POINT_SIZE, alpha=0.5, zorder = 0)
ax.set_xlabel(r"$d_a$", fontsize=LABEL_SIZE_3D, fontweight='bold', labelpad=20)
ax.set_ylabel(r"$T_{2a}$", fontsize=LABEL_SIZE_3D, fontweight='bold', labelpad=20)
ax.set_zlabel(r"$T_{2b}$", fontsize=LABEL_SIZE_3D, fontweight='bold', labelpad=20)
ax.tick_params(axis='both', which='major', labelsize=TICK_SIZE_3D)

for i, color in zip(selected_points, colors):
    ax.scatter(d_a[i], T_2a[i], T_2b[i], c=color, s=SELECTED_POINT_SIZE, marker = "D", edgecolors = "black", linewidth = 0.5)

ax.view_init(azim=-55, elev=30)

plt.tight_layout(pad = 1.8)
plt.gcf().set_size_inches(9, 8)

plt.savefig("results/plot1d.png")
plt.show()

plt.clf()
```

```python
# Subplot e

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d', computed_zorder=False)
ax.scatter(d_a, d_b, T_2a, s=POINT_SIZE, alpha=0.5, zorder = 0)
ax.set_xlabel(r"$d_a$", fontsize=LABEL_SIZE_3D, fontweight='bold', labelpad=20)
ax.set_ylabel(r"$d_b$", fontsize=LABEL_SIZE_3D, fontweight='bold', labelpad=20)
ax.set_zlabel(r"$T_{2a}$", fontsize=LABEL_SIZE_3D, fontweight='bold', labelpad=20)
ax.tick_params(axis='both', which='major', labelsize=TICK_SIZE_3D)

for i, color in zip(selected_points, colors):
    ax.scatter(d_a[i], d_b[i], T_2a[i], c=color, s=SELECTED_POINT_SIZE, marker = "D", edgecolors = "black", linewidth = 0.5)

ax.view_init(azim=40, elev=30)

plt.tight_layout(pad = 1.5)
plt.gcf().set_size_inches(9, 8)

plt.savefig("results/plot1e.png")
plt.show()

plt.clf()
```

```python
# Subplot f

TEs = np.arange(0, 401, 25)

plt.xlabel("TE", fontsize=LABEL_SIZE_2D)
plt.ylabel("Signal", fontsize=LABEL_SIZE_2D)

plt.xticks(fontsize=TICK_SIZE_2D)
plt.yticks(fontsize=TICK_SIZE_2D)

for i, color in zip(selected_points, colors):
    d_1, d_2, T_21, T_22 = data[i]
    signal = d_1*np.exp(-TEs/T_21) + d_2*np.exp(-TEs/T_22)
    plt.plot(TEs, signal, color = color, marker = "o", markersize = POINT_SIZE, alpha = 0.5, linewidth = 5)

plt.tight_layout(pad = 1)
plt.gcf().set_size_inches(8.5, 8)

# plt.savefig("results/plot1f.png")
plt.show()

plt.clf()
```
